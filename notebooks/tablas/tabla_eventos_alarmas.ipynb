```python
# Leer rutas

from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[1]
DATA_PATH = PROJECT_ROOT / "output"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("DATA_PATH:", DATA_PATH)
print("Existe data:", DATA_PATH.exists())
```

```python
# Excel con df_evento y df_alarma

import pandas as pd
from pathlib import Path

# -----------------------------------------------------
# 1) Leer los CSV ya generados
# -----------------------------------------------------
df_alarmas = pd.read_csv(PROJECT_ROOT / "output" /  "tablas" / "df_alarmas.csv")
df_eventos = pd.read_csv(PROJECT_ROOT / "output" /"tablas" / "df_eventos.csv")

# -----------------------------------------------------
# 2) Ordenar por maquina y fase (si existen)
# -----------------------------------------------------
if {'maquina', 'fase'}.issubset(df_alarmas.columns):
    df_alarmas = df_alarmas.sort_values(['maquina', 'fase'])

if {'maquina', 'fase'}.issubset(df_eventos.columns):
    df_eventos = df_eventos.sort_values(['maquina', 'fase'])

# -----------------------------------------------------
# 3) Preparar salida
# -----------------------------------------------------
OUT_DIR = PROJECT_ROOT / "output" / "tablas"
OUT_DIR.mkdir(parents=True, exist_ok=True)

output_file = OUT_DIR / "analisis_eventos_y_alarmas.xlsx"
# -----------------------------------------------------
# 4) Escribir Excel con dos hojas
# -----------------------------------------------------
with pd.ExcelWriter(output_file, engine='xlsxwriter') as writer:

    df_eventos.to_excel(
        writer,
        sheet_name='eventos',
        index=False
    )

    df_alarmas.to_excel(
        writer,
        sheet_name='alarmas',
        index=False
    )

print(f'Excel generado correctamente en: {output_file.resolve()}')
```
