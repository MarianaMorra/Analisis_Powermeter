```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
import numpy as np
```

```python
pio.renderers.default = 'browser'
```

```python
df = pd.read_csv(r"C:\Users\Diego\Desktop\provser\AnalizarSensores\CSVpreprocesados\S_240903_P2.csv")
```

```python
print(df.columns)
```

```python
fig = go.Figure()

modulo = np.sqrt(df["ax"]**2 + df["az"]**2)
modulo = modulo.interpolate()

# LÃ­nea de ax
fig.add_trace(go.Scatter(
    x=df["timestamp"],
    y=modulo,
    name="modulo",
    yaxis="y1",
    line=dict(color="blue")
))

fig.add_trace(go.Scatter(
    x=df["timestamp"],
    y=df["velocidad"],
    name="Velocidad",
    yaxis="y2",
    line=dict(color="red"),
    customdata=df[["Unnamed: 0"]],
    hovertemplate=(
        "Tiempo: %{x}<br>" +
        "Velocidad: %{y}<br>" +
        "ID: %{customdata[0]}<extra></extra>"
    )
))

# Layout con dos ejes
fig.update_layout(
    title="Velocidad y Modulo vs Tiempo",
    xaxis=dict(title="Tiempo"),
    yaxis=dict(title="Modulo", side="left"),
    yaxis2=dict(title="Velocidad", overlaying="y", side="right"),
    legend=dict(x=0.01, y=0.99),
    hovermode="x unified"
)

fig.show()
```
